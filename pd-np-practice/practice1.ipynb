```python
import numpy as np
reshape_mat = np.array(np.arange(30)).reshape((5,6))
print(reshape_mat)
```

```python
mat = np.array([[1, 2, 3], [4, 5, 6]])
print(mat)
slicing_mat = mat[0,:]
print(slicing_mat)

slicing_mat = mat[:, 1]
print(slicing_mat)
```

```python
mat = np.array([[1, 2], [4, 5], [7, 8]])
print(mat)
print(mat[1,0])
print(mat[2, 1], mat[0, 1])

# mat[[2행, 1행],[0열, 1열]]
indexing_mat = mat[[2, 1],[0, 1]]
# 2, 0 -> 7이 되고, 1, 1 -> 5가 됨.
print(indexing_mat)
```

```python
x = np.array([1,2,3])
y = np.array([4,5,6])

# result = np.add(x, y)와 동일
# result = np.subtract(x, y)와 동일.
# result = np.multiply(x, y)와 동일.
# result = np.divide(x, y)와 동일.
print(x+y, x-y, x*y, x/y)
```

```python
#행렬 곱
mat1 = np.array([[1,2],[3,4]])
mat2 = np.array([[5,6],[7,8]])
mat3 = np.dot(mat1, mat2)
print(mat1)
print(mat2)
print(mat3)
print(mat1 @ mat2)
```

```python
import matplotlib.pyplot as plt

plt.title('test1')
plt.plot([1,2,3,4],[2,4,8,6])
plt.show()

plt.title('test2')
plt.plot([1,2,3,4],[2,4,8,6])
plt.xlabel('hours')
plt.ylabel('score')
plt.plot([1.5,2.5,3.5,4.5],[3,5,8,10]) # 라인 새로 추가
plt.legend(['A student', 'B student']) # 범례 삽입
plt.show()
```

```python
a = np.array([[1, 2], [3, 4]])
b = np.array([[5, 6]])

# vstack 사용
result = np.vstack((a, b))
print(result)

# concatenate 사용 (axis=0)
result = np.concatenate((a, b), axis=0)
print(result, "\n\n")


a = np.array([[1, 2], [3, 4], [5, 6]])
b = np.array([[7, 8], [9, 10], [10, 11]])

# hstack 사용
result = np.hstack((a, b))
print(result)

# concatenate 사용 (axis=1)
result = np.concatenate((a, b), axis=1)
print(result)

# 3차원 이상의 배열은 concatenate 사용 (axis=[해당 축])
```
